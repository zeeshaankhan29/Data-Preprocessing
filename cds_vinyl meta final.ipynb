### Importing Libraries and loading Data

```python
import pandas as pd
import gzip
import json
import re
import numpy as np

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df1 = getDF("E:\project\meta_CDs_and_Vinyl.json.gz")
```

# Price Column Cleaning

```python
price = []
for i in df['price']:
    if len(i) < 15:
        price.append(i)
    else:
        i = 0
        price.append(i)
        
df['price'] = pd.DataFrame(price)
df['price']
```

```python
import re
```

```python
df['price'] = df['price'].apply(lambda x : str(x).replace("$",""))

df['price'] = df['price'].apply(lambda x : str(x).replace(',',''))

df.price[df.price.str.contains("-")]=df.price[df.price.str.contains("-")].apply(lambda i : i.replace("-",'')[0:5])

df['price'] = df['price'].replace("",0)

df['price'] = df['price'].replace("\n\n\n<script",0)

df['price'] = df['price'].apply(lambda i : re.sub('/D','',str(i)))

df['price'] = pd.to_numeric(df['price'], downcast="float")

df['price'] = df['price'].replace(0,np.nan)
```

# Rank Column Cleaning 

## Extracting numeric rank 

```python
a = df['rank']

rank = []

rank_cat = []

lst = []

for i in a:
    i = str(i)
    i = i.split(" in ")[0]
    i = i.replace(',','')
    i = i.replace("['>#",'')
    rank.append(i)
    
for i in a:
    i = str(i)
    i = i.split(" in ")[-1]
    i = i.replace(',','')
    i = i.replace(' (','')
    lst.append(i)

for i in lst:
    i = str(i)
    i = i.split(" >")[0]
    i = i.split(" &gt")[0]
    i = i.replace("See top 100)']","")
    i = i.replace(")']","")
    i = i.replace("&amp;","")
    i = i.replace("&","")
    rank_cat.append(i)
```

```python
rank_cat = pd.DataFrame(rank_cat)
rank_cat.rename(columns = {0:'cat'}, inplace = True)
rank_cat.cat = rank_cat.cat.replace("[]",np.nan)
rank_cat.value_counts()
```

```python
df['rank_cat'] = rank_cat
df['rank'] = rank
df.head()
```
